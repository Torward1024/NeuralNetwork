## <u>Введение</u>

Целью проекта является внедрение методов машинного обучения для астрономической обсерватории, чтобы определить поверхностную температуру наблюдаемых звезд. Для расчёта температуры используются следующие ключевые законы физики и методы: закон смещения Вина, закон Стефана-Больцмана, методики спектрального анализа.
На момент выполнения проекта, в каталоге обсерватории находятся характеристики уже изученных 240 звёзд.

Для достижения поставленной цели необходимо решить следующие задачи:
1. Выполнить исследовательский анализ имеющихся данных:
 - По результату исследовательского анализа внести корректировки, если они необходимы;
 - При необходимости выполнить категоризацию исходных данных;
 - Подготовить обучающую и тестовую выборки;
 - Выполнить масштабирование количественных данных; 
2. Разработать решение на базе нейронной сети для предсказания температуры:
 - Построить простую модель нейронной сети и выполнить её обучение;
 - По результатам обучения выполнить улучшение сети;
 
**Описание данных**
В данных находится следующая информация о 240 наблюдавшихся звёздах:
- Относительная светимость $L/L_o$ -- светимость звезды относительно Солнца.
- Относительный радиус $R/R_o$ -- радиус звезды относительно радиуса Солнца.
- Абсолютная звёздная величина $M_\nu$ -- физическая величина, характеризующая блеск звезды.
- Звёздный цвет (```white```, ```red```, ```blue```, ```yellow```, ```yellow-orange``` и др.) -- цвет звезды, который определяют на основе спектрального анализа.
- Тип звезды.
|Тип звезды|Номер, соответствующий типу|
|:---|---:|
|Коричневый карлик|	0|
|Красный карлик|	1|
|Белый карлик|	2|
|Звёзды главной последовательности|	3|
|Сверхгигант|	4|
|Гипергигант|	5|
- Абсолютная температура T(K) — температура на поверхности звезды в Кельвинах.

**Дополнительная информация**
- Светимость Солнца (англ. Average Luminosity of Sun): $L_{\odot}=3.828\cdot10^{26}$ Вт
- Радиус Солнца (англ. Average Radius of Sun)  $R_{\odot}=6.9551\cdot10^8$ м

## <u>Класс нейросети</u>

**Комментарий**: ниже представлен класс нейросети для дальнейшей работы с ней.

````python
class NNet(nn.Module):
    """
    Универсальный класс нейронной сети на базе nn.Sequential с поддержкой BatchNorm.
    
    Класс создаёт полносвязную нейронную сеть с гибкой настройкой архитектуры,
    ф-ций активации, инициализации весов и метода обучения.
    
    Применение:
    ----------
    - Регрессия с произвольной архитектурой
    - Классификация с различными выходными активациями
    - Эксперименты с различными инициализациями и нормализациями
    
    Пример использования:
    net = NNet([10, 64, 32, 1], activation='ReLU', dropout_rate=0.2, 
               use_batchnorm=True)
    """
    
    def __init__(self, n_neurons, activation='ReLU', dropout_rate=0.0, 
                 weight_init='kaiming', bias_init='zeros', output_activation=None,
                 activation_params=None, use_batchnorm=False, bn_momentum=0.1):
        """
        Инициализация универсальной нейронной сети.
        
        Параметры:
        ----------
        n_neurons -- список целых чисел, определяющий архитектуру сети
                     (первый элемент - размер входа, последний - выход)
        activation -- строка, тип активационной функции для скрытых слоёв
                      возможные варианты: 'ReLU', 'LeakyReLU', 'ELU', 'SiLU', 
                      'Tanh', 'Sigmoid'
        dropout_rate -- вещественное число, вероятность dropout (0.0 = без dropout)
        weight_init -- строка, метод инициализации весов
                       варианты: 'normal', 'xavier', 'kaiming'
        bias_init -- строка, метод инициализации смещений
                     варианты: 'normal', 'zeros', 'constant'
        output_activation -- строка или None, активация для выходного слоя
                             варианты: 'Sigmoid', 'Softmax', None
        activation_params -- словарь или None, параметры активаций
                             например: {'alpha': 0.01} для LeakyReLU
        use_batchnorm -- булево, использовать ли BatchNorm1d после линейных слоёв
        bn_momentum -- вещественное число, момент для Batch Normalization (по умолчанию 0.1)
        
        Примечания:
        -----------
        - BatchNorm и активации применяются только к скрытым слоям
        - Для регрессии обычно output_activation=None
        - Dropout применяется после активаций скрытых слоёв
        - При batch_size <=0: обучение на полном датасете за один проход на эпоху (без DataLoader)
        """
        # базовая инициализация родительского класса
        super(NNet, self).__init__()
        
        # проверка и установка параметров активации по умолчанию
        if activation_params is None:
            activation_params = {}
        
        # список для хранения всех слоёв сети
        layers = []
        
        # основной цикл построения архитектуры
        for i in range(len(n_neurons) - 1):
            # добавление линейного слоя
            layers.append(nn.Linear(n_neurons[i], n_neurons[i+1]))
            
            # BatchNorm после линейного слоя для скрытых слоёв (не для последнего)
            if i < len(n_neurons) - 2 and use_batchnorm:
                layers.append(nn.BatchNorm1d(n_neurons[i+1], momentum=bn_momentum))
            
            # активация и dropout для скрытых слоёв (не для последнего)
            if i < len(n_neurons) - 2:
                # выбор активационной функции
                if activation == 'ReLU':
                    layers.append(nn.ReLU())
                elif activation == 'LeakyReLU':
                    alpha = activation_params.get('alpha', 0.01)
                    layers.append(nn.LeakyReLU(negative_slope=alpha))
                elif activation == 'ELU':
                    alpha = activation_params.get('alpha', 1.0)
                    layers.append(nn.ELU(alpha=alpha))
                elif activation == 'SiLU':  # Swish активация
                    layers.append(nn.SiLU())
                elif activation == 'Tanh':
                    layers.append(nn.Tanh())
                elif activation == 'Sigmoid':
                    layers.append(nn.Sigmoid())
                else:
                    # обработка неподдерживаемой активации
                    raise ValueError(f"Неподдерживаемая активация: {activation}")
                
                # добавление dropout для скрытых слоёв при необходимости
                if dropout_rate > 0:
                    layers.append(nn.Dropout(p=dropout_rate))
            else:
                # активация выходного слоя (если указана)
                if output_activation == 'Sigmoid':
                    layers.append(nn.Sigmoid())
                elif output_activation == 'Softmax':
                    layers.append(nn.Softmax(dim=1))
                # для регрессии обычно без выходной активации
        
        # создаем модель из всех слоёв
        self.model = nn.Sequential(*layers)
        
        # инициализация весов с учётом ф-ции активации
        self._init_weights(weight_init, bias_init, activation)
        
        # сохранение параметров
        self.use_batchnorm = use_batchnorm
        self.dropout_rate = dropout_rate
        self.activation = activation
    
    def _init_weights(self, weight_init, bias_init, activation):
        """
        Инициализация весов и смещений с учётом типа нелинейности.
        
        Параметры:
        ----------
        weight_init -- метод инициализации весов
        bias_init -- метод инициализации смещений  
        activation -- тип ф-ции активации для выбора подходящей инициализации
        
        Логика:
        -------
        - Для ReLU-подобных функций используется kaiming инициализация
        - Для Tanh/Sigmoid - xavier или стандартная нормальная
        - Bias обычно инициализируется нулями для стабильности
        """
        def init_fn(layer):
            # применение только к линейным слоям
            if isinstance(layer, nn.Linear):
                # выбор нелинейности для kaiming инициализации
                nonlinearity = 'relu'  # значение по умолчанию
                if activation in ['Tanh', 'Sigmoid']:
                    nonlinearity = 'tanh' if activation == 'Tanh' else 'sigmoid'
                elif activation == 'LeakyReLU':
                    nonlinearity = 'leaky_relu'
                
                # инициализация весов
                if weight_init == 'normal':
                    # нормальное распределение с малой дисперсией
                    nn.init.normal_(layer.weight, mean=0.0, std=0.02)
                elif weight_init == 'xavier':
                    # Xavier/Glorot uniform инициализация
                    nn.init.xavier_uniform_(layer.weight)
                elif weight_init == 'kaiming':
                    # He инициализация с учётом нелинейности
                    nn.init.kaiming_normal_(layer.weight, nonlinearity=nonlinearity)
                else:
                    raise ValueError(f"Неподдерживаемая инициализация весов: {weight_init}")
                
                # инициализация смещений
                if bias_init == 'normal':
                    nn.init.normal_(layer.bias, mean=0.0, std=0.02)
                elif bias_init == 'zeros':
                    # нули - наиболее стабильный выбор
                    nn.init.zeros_(layer.bias)
                elif bias_init == 'constant':
                    # постоянное значение (по умолчанию 0.1)
                    nn.init.constant_(layer.bias, 0.1)
                else:
                    raise ValueError(f"Неподдерживаемая инициализация смещений: {bias_init}")
        
        # применение функции инициализации ко всем слоям модели
        self.model.apply(init_fn)
    
    def forward(self, x):
        """
        Прямой проход через сеть.
        
        Параметры:
        ----------
        x -- входные данные тензор размерности [batch_size, input_size]
        
        Возвращает:
        ----------
        выход сети в виде одномерного тензора [batch_size * output_size]
        
        Примечание:
        -----------
        .flatten() используется для упрощения работы с регрессионными задачами,
        где выход часто одномерный
        """
        # применение модели и преобразование в одномерный формат
        return self.model(x).flatten()
    
    def train_model(self, X_train, y_train, X_test, y_test, 
                learning_rate=1e-3, num_epochs=15000, batch_size=32,
                loss_fn=None, optimizer_type='Adam', print_every=500):
        """
        Обучение нейронной сети с мини-батчами и мониторингом метрик.

        Параметры:
        ----------
        X_train : torch.Tensor
            Обучающий набор признаков
        y_train : torch.Tensor
            Обучающий набор целевых значений
        X_test : torch.Tensor
            Тестовый набор признаков
        y_test : torch.Tensor
            Тестовый набор целевых значений
        learning_rate : float, optional
            Скорость обучения оптимизатора (по умолчанию 1e-3)
        num_epochs : int, optional
            Количество эпох обучения (по умолчанию 15000)
        batch_size : int, optional
            Размер мини-батча (если <=0, используется полный батч, по умолчанию 32)
        loss_fn : torch.nn.Module, optional
            Функция потерь (по умолчанию MSELoss для регрессии)
        optimizer_type : str, optional
            Тип оптимизатора: 'Adam', 'SGD', 'RMSprop' (по умолчанию 'Adam')
        print_every : int, optional
            Интервал вывода статистики обучения (в эпохах, по умолчанию 500)

        Возвращает:
        -----------
        history_df : pd.DataFrame
            DataFrame с историей обучения, содержит столбцы:
            - epoch: номер эпохи
            - loss: значение функции потерь на обучающем наборе
            - rmse_train: RMSE на обучающем наборе
            - rmse_test: RMSE на тестовом наборе

        Особенности:
        ------------
        - Автоматическое переключение между режимами train/eval для BatchNorm и Dropout
        - Вычисление RMSE на обучающей и тестовой выборках
        - Поддержка различных оптимизаторов
        - При batch_size <=0: обучение на полном датасете за один проход на эпоху (без DataLoader)
        """
        # настройка функции потерь (MSE по умолчанию для регрессии)
        if loss_fn is None:
            loss_fn = nn.MSELoss()

        # выбор и настройка оптимизатора
        if optimizer_type == 'Adam':
            optimizer = torch.optim.Adam(self.parameters(), lr=learning_rate)
        elif optimizer_type == 'SGD':
            optimizer = torch.optim.SGD(self.parameters(), lr=learning_rate)
        elif optimizer_type == 'RMSprop':
            optimizer = torch.optim.RMSprop(self.parameters(), lr=learning_rate)
        else:
            raise ValueError(f"Неподдерживаемый оптимизатор: {optimizer_type}")

        # список для хранения истории обучения
        history_list = []

        # определение режима обучения: мини-батчи или полный батч
        use_full_batch = (batch_size <= 0)
        if use_full_batch:
            num_batches = 1
        else:
            train_dataset = TensorDataset(X_train, y_train)
            train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
            num_batches = len(train_loader)

        # основной цикл обучения
        for epoch in range(num_epochs):
            # переключение в режим обучения (важно для BatchNorm и Dropout)
            self.train()
            total_loss = 0

            # выбор логики обучения в зависимости от режима
            if use_full_batch:
                batch_X = X_train
                batch_y = y_train
                optimizer.zero_grad()
                preds = self.forward(batch_X)
                loss_value = loss_fn(preds, batch_y.flatten())
                loss_value.backward()
                optimizer.step()
                total_loss = loss_value.item()
            else:
                for batch_X, batch_y in train_loader:
                    optimizer.zero_grad()
                    preds = self.forward(batch_X)
                    loss_value = loss_fn(preds, batch_y.flatten())
                    loss_value.backward()
                    optimizer.step()
                    total_loss += loss_value.item()

            # вычисление метрик на каждой эпохе
            self.eval()
            with torch.no_grad():
                # метрики на обучающей выборке
                train_preds = self.forward(X_train).flatten()
                train_rmse = torch.sqrt(loss_fn(train_preds, y_train.flatten()))

                # метрики на тестовой выборке
                test_preds = self.forward(X_test).flatten()
                test_rmse = torch.sqrt(loss_fn(test_preds, y_test.flatten()))

            # вычисление средней ошибки по эпохе
            avg_loss = total_loss / num_batches

            # сохранение метрик в историю
            history_list.append({
                'epoch': epoch + 1,
                'loss': avg_loss,
                'rmse_train': train_rmse.item(),
                'rmse_test': test_rmse.item()
            })

            # вывод прогресса
            if epoch % print_every == 0 or epoch == num_epochs - 1:
                bn_status = "BN" if self.use_batchnorm else "No BN"
                print(f'Эпоха: {epoch:4d}, loss: {avg_loss:.6f}, '
                      f'RMSE (train): {train_rmse:.4f}, RMSE (test): {test_rmse:.4f} [{bn_status}]')

        # создание DataFrame из истории
        history_df = pd.DataFrame(history_list)

        # финальное переключение в режим оценки
        self.eval()
        return history_df
    
    def predict(self, X):
        """
        Получение предсказаний для новых данных.
        
        Параметры:
        ----------
        X -- входные данные для предсказания (тензор)
        
        Возвращает:
        ----------
        предсказания в виде одномерного тензора
        
        Примечание:
        -----------
        Автоматическое переключение в режим оценки и отключение градиентов
        """
        # обеспечение режима оценки
        self.eval()
        
        with torch.no_grad():
            return self.forward(X).flatten()
    
    def get_metrics(self, X, y, loss_fn=None):
        """
        Вычисление метрик качества модели (MSE и RMSE).
        
        Параметры:
        ----------
        X -- входные данные (тензор)
        y -- истинные значения (тензор)
        loss_fn -- функция потерь (по умолчанию MSELoss)
        
        Возвращает:
        ----------
        словарь с метриками: {'MSE': значение, 'RMSE': значение}
        """
        # функция потерь по умолчанию
        if loss_fn is None:
            loss_fn = nn.MSELoss()
        
        # режим оценки
        self.eval()
        with torch.no_grad():
            # получаем предсказания
            preds = self.predict(X)
            
            # выполняем вычисление метрик
            mse = loss_fn(preds, y.flatten())
            rmse = torch.sqrt(mse)
            
            # возвращаем результаты как словарь
            return {'MSE': mse.item(), 'RMSE': rmse.item()}
    
    def get_architecture_info(self):
        """
        Вывод информации об архитектуре сети.
        
        Возвращает:
        ----------
        строка с описанием архитектуры, нормализации, dropout и активации
        
        Пример вывода:
        --------------
        "Архитектура: Linear(in_features=10, out_features=64) -> ... (линейные слои)
        BatchNorm: Да (momentum=0.1)
        Dropout: 0.2
        Размер батча: 16
        Activation: ReLU"
        """
        # получаем информацию о линейных слоях (каждый второй элемент)
        linear_layers = self.model[::2] if len(self.model) > 1 else [self.model[0]]
        info = f"Архитектура: {' -> '.join(map(str, linear_layers))} (линейный слои)"
        
        # добавляем информацию о нормализации
        if self.use_batchnorm:
            info += f"\nBatchNorm: Да (момент = {0.1})"
        
        # добавляем информацию о dropout
        if self.dropout_rate > 0:
            info += f"\nDropout: {self.dropout_rate}"
        
        # добавляем информации о ф-ции активации
        info += f"\nФ-ция активации: {self.activation}"
        
        return info
````

````python
# функция для отображения количества пропусков в процентном соотношении от общего числа записей
def plot_isna(data):
    try:
        # вычисляем процент пропусков
        missing_data = data.isna().mean() * 100
        # фильтруем столбцы с пропусками и сортируем
        missing_data = missing_data[missing_data > 0].sort_values(ascending=True)
        
        if missing_data.empty:
            print('Пропуски в данных отсутствуют.')
            return
        
        # создаем горизонтальную столбчатую диаграмму
        fig, ax = plt.subplots(figsize=(13, 6))
        ax.barh(missing_data.index, missing_data, color='skyblue', edgecolor='black')
        
        # Настраиваем подписи и заголовок
        ax.set_title('% пропусков в данных по столбцам\n(от общего количества записей в данных)', fontsize=14)
        ax.set_xlabel('Пропуски в данных, (%)', fontsize=12)
        ax.set_ylabel('Столбцы', fontsize=12)
        
        # Настраиваем шрифт для меток на осях
        ax.tick_params(axis='both', labelsize=12)
        
        # Добавляем сетку
        ax.grid(True, axis='x', linestyle='--', alpha=0.7)
        
        # Отображаем график
        plt.show()
        print("Доля пропусков по столбцам:")
        display(missing_data)
        
    except Exception as e:
        print(f'Ошибка: {e}')
````

### Функция ```plot_data_analysis```

````python
# вспомогательная функция для отрисовки параметров с целью их дальнейшего анализа
# функция была написана ранее в рамках предыдущих проектов, эта версия доработанная
def plot_data_analysis(
    data,
    title: str = 'Изучение параметров',
    title_box: str = 'Диаграмма размаха',
    title_hist: str = 'Распределение',
    x_label: str = 'X',
    y_label: str = 'Y',
    plot_box: bool = True,
    plot_bar: bool = False,
    x=None,
    y=None,
    bins: int = 200,
    bar_labels=None,
    discrete=False,
    figsize=None,
    color: str = 'skyblue',
    show_stats: bool = True,
    label_angle: int = 0
):
    """
    Функция для отрисовки данных с возможностью построения:
        boxplot (диаграмма размаха)
        гистограммы или barplot (столбчатой диаграммы)
    
    Параметры:
    ----------
        data -- данные для анализа (pd.Series, pd.DataFrame или массив)
        title -- общий заголовок
        title_box -- заголовок для boxplot
        title_hist -- заголовок для гистограммы/barplot
        x_label -- подпись оси X
        y_label -- подпись оси Y
        plot_box -- строить ли boxplot (по умолчанию True)
        plot_bar -- использовать ли barplot вместо гистограммы (по умолчанию False)
        x -- данные для оси X (если нужен barplot с внешними данными)
        y -- данные для оси Y (если нужен barplot с внешними данными)
        bins -- количество бинов для гистограммы
        bar_labels -- подписи для barplot
        discrete -- флаг для дискретных данных
        figsize -- размер графика
        color -- основной цвет графиков
        show_stats -- показывать ли статистику (по умолчанию True)
    """
    # проверка и преобразование типов ходных данных
    if not isinstance(data, (pd.Series, pd.DataFrame, np.ndarray, list)):
        raise TypeError("Данные должны быть типа pd.Series, pd.DataFrame, np.ndarray или list")
    
    if isinstance(data, (pd.DataFrame, np.ndarray, list)):
        data = pd.Series(data)
    
    # настройка размера графика
    if figsize is None:
        figsize = (13, 8) if plot_box else (13, 6)
    
    # создание subplots
    if plot_box:
        fig, (ax_box, ax_main) = plt.subplots(2, 1, figsize=figsize, 
                                             gridspec_kw={'height_ratios': [1, 2]})
        # усы
        ax_box.boxplot(data, vert=False, patch_artist=True,
                      boxprops=dict(facecolor='lightblue'))
        ax_box.set_title(title_box, pad=10)
        ax_box.set_xlabel(x_label)
        ax_box.grid(axis='x', linestyle='--', alpha=0.7)
        ax_box.set_yticks([])
    else:
        fig, ax_main = plt.subplots(1, 1, figsize=figsize)
    
    # выбор типа основного графика
    if plot_bar:
        if x is None or y is None:
            if discrete:
                value_counts = data.value_counts().sort_index()
                x = value_counts.index
                y = value_counts.values
                bar_labels = x if bar_labels is None else bar_labels
            else:
                raise ValueError("Для barplot нужно указать x и y или использовать discrete=True")
        
        bars = ax_main.bar(x=x, height=y, color=color, edgecolor='black', alpha=0.8)
        ax_main.set_title(title_hist, pad=10)
        
        # установка подписей для дискретных данных
        if bar_labels is not None:
            ax_main.set_xticks(x if discrete else np.arange(len(x)))
            ax_main.set_xticklabels(bar_labels, rotation=label_angle, ha='right')
    else:
        # гистограмма
        ax_main.hist(data, bins=bins, color=color, edgecolor='black', alpha=0.8)
        ax_main.set_title(title_hist, pad=10)
    
    # подписи к осям и сетка
    ax_main.set_xlabel(x_label)
    ax_main.set_ylabel(y_label)
    ax_main.grid(axis='both', linestyle='--', alpha=0.5)
    
    fig.suptitle(title, fontsize=14, y=1.02)
    plt.tight_layout()
    plt.show()
    
    # вывод статистики
    if show_stats:
        display(data.describe())
````

### Функция ```get_star_color```

````python
def get_star_color(temperature):
    """
    Функция для отрисовки данных с возможностью построения:
        boxplot (диаграмма размаха)
        гистограммы или barplot (столбчатой диаграммы)
    
    Параметры:
    ----------
    temperature -- температура поверхности звезды в К
    
    Возвращает:
    ----------- 
    цвет звезд в зависимости от температуры (str) -- blue, blue white,
                                                     white, yellow white, 
                                                     yellow, orange, red
    """
    if temperature >= 28000:
        return 'blue'
    elif temperature >= 10000 and temperature < 28000:
        return 'blue white'
    elif temperature >= 7000 and temperature < 10000:
        return 'white'
    elif temperature >= 6000 and temperature < 7000:
        return 'yellow white'
    elif temperature >= 5000 and temperature < 6000:
        return 'yellow'
    elif temperature >= 3500 and temperature < 5000:
        return 'orange'
    else:
        return 'red'
````

### Функция ```get_mag_cat```

````python
def get_mag_cat(mag):
    """
    Функция для преобразования абсолютной звездной величины в категории яркости
    Параметры:
    ----------
    mag -- значение абсолютной звездной величины звезды
    
    Возвращает:
    ----------- 
    категория яркости: ultra_bright, bright, medium_bright, moderate, faint, very_faint
    """
    if mag < -5:
        return 'ultra_bright'
    elif mag < -2:
        return 'bright'
    elif mag < 2:
        return 'medium_bright'
    elif mag < 6:
        return 'moderate'
    elif mag < 10:
        return 'faint'
    else:
        return 'very_faint'
````

### Функция ```plot_corr_heatmap```

````python
# функция построения тепловой карты коэффициентов корреляции Пирсона
def plot_corr_heatmap(data, 
                      title: str = 'Тепловая карта корреляции', 
                      columns: [str] = None):
    """
    Функция для построения тепловой карты коэффициентов корреляции Пирсона
    
    Параметры:
    ----------
        data -- DataFrame
    """
    plt.figure(figsize=(14, 6))
    corr_matrix = data.phik_matrix(interval_cols=columns)
    sns.heatmap(corr_matrix, 
                annot=True,  # показывать значения в ячейках
                fmt=".2f",   # формат чисел (2 знака после запятой)
                cmap='coolwarm',  # цветовая схема
                vmin=-1, vmax=1,  # диапазон значений
                linewidths=0.5)   # ширина линий между ячейками

    # заголовок
    plt.title(title)

    # вывод графика
    plt.tight_layout()
    plt.show()
    return corr_matrix
````

### Функция ```corr_analysis```

````python
def corr_analysis(corr_matrix, 
                  target_feature, 
                  features=None, 
                  min_abs_corr=0.1):
    """
    Анализирует корреляции целевого признака с другими признаками и сортирует их по убыванию.
    
    Параметры:
    ----------
    corr_matrix : DataFrame
        Матрица корреляции (phik-матрица или обычная корреляционная матрица)
    target_feature : str
        Название целевого признака для анализа
    features : list, optional
        Список признаков для анализа (если None, берутся все из матрицы)
    min_abs_corr : float, optional
        Минимальная абсолютная корреляция для включения в результат (по умолчанию 0.1)
        
    Возвращает:
    -----------
    DataFrame
        Таблица с признаками, отсортированными по убыванию абсолютной корреляции с целевым признаком,
        с указанием силы связи по шкале Чеддока
    """
    # если не указан список признаков, берем все из матрицы (исключая целевой)
    if features is None:
        features = [col for col in corr_matrix.columns if col != target_feature]
    
    # собираем корреляции с целевым признаком
    correlations = []
    
    for feature in features:
        if feature == target_feature:
            continue
        
        corr = corr_matrix.loc[target_feature, feature]
        abs_corr = abs(corr)
        
        # реализуем шкалу Чеддока (согласно 11 спринту - 3 тема - 9 урок)
        if abs_corr >= 0.9:
            strength = 'Весьма высокая'
        elif abs_corr >= 0.7:
            strength = 'Высокая'
        elif abs_corr >= 0.5:
            strength = 'Заметная'
        elif abs_corr >= 0.3:
            strength = 'Умеренная'
        elif abs_corr >= 0.1:
            strength = 'Слабая'
        else:
            continue  # пропускаем если корреляция ниже min_abs_corr
        
        # определяем направление связи
        direction = 'положительная' if corr > 0 else 'отрицательная'
        
        correlations.append({
            'Признак': feature,
            'Корреляция': corr,
            'Абс. корреляция': abs_corr,
            'Сила связи': strength,
            'Направление': direction
        })
    
    # создаем DataFrame и сортируем по убыванию абсолютной корреляции
    if not correlations:
        return pd.DataFrame()  # возвращаем пустой DataFrame если нет корреляций
    
    result_df = pd.DataFrame(correlations).sort_values(
        by='Абс. корреляция', ascending=False
    ).reset_index(drop=True)
    
    # вывод корреляции для удобства чтения
    result_df['Корреляция'] = result_df['Корреляция'].apply(lambda x: f"{x:.3f}")
    
    return result_df[['Признак', 'Корреляция', 'Сила связи', 'Направление']]
````

### Функция ```plot_fact_forecast```

````python
def plot_fact_forecast(
    y_true,
    y_pred,
    title: str = 'Сравнение факта, прогноза и остатков',
    x_label: str = 'Наблюдения',
    y_label: str = 'Значения',
    residual_label: str = 'Остатки (прогноз - факт)',
    figsize: tuple = (13, 6),
    fact_alpha: float = 0.7,
    forecast_alpha: float = 0.5,
    residual_alpha: float = 0.7,
    show_stats: bool = True,
    label_angle: int = 0
):
    """
    Функция для визуального сравнения фактических и прогнозных значений
    в виде столбчатой диаграммы с наложением прогноза на факт и линии остатков.

    Параметры:
    ----------
    y_true : array-like
        Фактические значения (pd.Series, np.ndarray, list)
    y_pred : array-like
        Прогнозные значения (pd.Series, pd.DataFrame, np.ndarray, list)
    title : str, optional
        Заголовок графика (по умолчанию 'Сравнение факта, прогноза и остатков')
    x_label : str, optional
        Подпись оси X (по умолчанию 'Наблюдения')
    y_label : str, optional
        Подпись оси Y для фактических и прогнозных значений (по умолчанию 'Значения')
    residual_label : str, optional
        Подпись оси Y для остатков (по умолчанию 'Остатки (прогноз - факт)')
    figsize : tuple, optional
        Размер графика (по умолчанию (13, 6))
    fact_alpha : float, optional
        Прозрачность столбцов факта (по умолчанию 0.7)
    forecast_alpha : float, optional
        Прозрачность столбцов прогноза (по умолчанию 0.5)
    residual_alpha : float, optional
        Прозрачность линии остатков (по умолчанию 0.7)
    show_stats : bool, optional
        Показывать ли статистику сравнения (по умолчанию True)
    label_angle : int, optional
        Угол наклона подписей по оси X (по умолчанию 0)

    Примечания:
    -----------
    - График отображает фактические и прогнозные значения как столбцы, а остатки (y_pred - y_true) как линию.
    - Остатки отображаются на вторичной оси Y для учета возможных различий в масштабе.
    - При show_stats выводится статистика по фактическим, прогнозным значениям и остаткам.
    """
    if not isinstance(y_true, (pd.Series, pd.DataFrame, np.ndarray, list)):
        raise TypeError("y_true должен быть типа pd.Series, pd.DataFrame, np.ndarray или list")

    if not isinstance(y_pred, (pd.Series, pd.DataFrame, np.ndarray, list)):
        raise TypeError("y_pred должен быть типа pd.Series, pd.DataFrame, np.ndarray или list")

    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    if len(y_true) != len(y_pred):
        raise ValueError("Размеры y_true и y_pred должны совпадать")

    n_observations = len(y_true)
    x_index = np.arange(n_observations)
    residuals = y_pred - y_true

    fig, ax1 = plt.subplots(figsize=figsize)

    # строим столбцы для факта
    ax1.bar(x_index, y_true, label='Факт', alpha=fact_alpha, color='skyblue',
            edgecolor='navy', linewidth=1.2)

    # строим столбцы для прогноза
    ax1.bar(x_index, y_pred, label='Прогноз', alpha=forecast_alpha, color='lightcoral',
            edgecolor='darkred', linewidth=1.2)

    # настраиваем основную ось Y
    ax1.set_xlabel(x_label)
    ax1.set_ylabel(y_label)
    ax1.set_title(title, fontsize=14, pad=20)

    # создаем вторичную ось Y для остатков
    ax2 = ax1.twinx()
    ax2.plot(x_index, residuals, label='Остатки', alpha=residual_alpha, color='limegreen',
             linewidth=2, marker='^', markersize=4)
    ax2.set_ylabel(residual_label)

    # объединяем легенды обеих осей
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='best')

    # настраиваем ось X
    if n_observations > 30:
        step = max(1, n_observations // 20)
        ax1.set_xticks(x_index[::step])
        ax1.set_xticklabels(x_index[::step], rotation=label_angle)
    else:
        ax1.set_xticks(x_index)
        ax1.set_xticklabels(x_index, rotation=label_angle)

    # сеточку
    ax1.grid(axis='y', linestyle='--', alpha=0.7)
    ax1.set_axisbelow(True)

    # настройка пределов для основной оси
    y_min = min(y_true.min(), y_pred.min())
    y_max = max(y_true.max(), y_pred.max())
    y_range = y_max - y_min
    ax1.set_ylim(y_min - 0.1 * y_range, y_max + 0.1 * y_range)

    # настройка пределов для вторичной оси
    res_min = residuals.min()
    res_max = residuals.max()
    res_range = res_max - res_min
    ax2.set_ylim(res_min - 0.1 * res_range, res_max + 0.1 * res_range)

    plt.tight_layout()
    plt.show()

    # статистика
    if show_stats:
        stats_df = pd.DataFrame({
            'Факт': y_true,
            'Прогноз': y_pred,
            'Остатки': residuals,
            'Абс. ошибка': np.abs(residuals)
        })

        display(stats_df.describe())

        # доп метрики
        mae = np.mean(np.abs(residuals))
        mse = np.mean(residuals ** 2)
        rmse = np.sqrt(mse)
        mape = np.mean(np.abs(residuals / y_true)) * 100 if np.all(y_true != 0) else np.nan

        print(f"RMSE: {rmse:.4f}")
````

### Функция ```plot_training_history```

````python
def plot_training_history(
    history_df,
    title: str = 'Динамика RMSE и остатков во время обучения',
    x_label: str = 'Эпоха',
    y_label: str = 'RMSE',
    figsize: tuple = (13, 6),
    train_alpha: float = 0.9,
    test_alpha: float = 0.9,
    residual_alpha: float = 0.9,
    show_stats: bool = True,
    label_angle: int = 0
):
    """
    Функция для визуализации RMSE на обучающей и тестовой выборках, а также остатков
    (разницы RMSE_test - RMSE_train) в виде линейного графика по эпохам.

    Параметры:
    ----------
    history_df : pd.DataFrame
        DataFrame с историей обучения, содержащий столбцы:
        - epoch: номер эпохи
        - rmse_train: RMSE на обучающем наборе
        - rmse_test: RMSE на тестовом наборе
    title : str, optional
        Заголовок графика (по умолчанию 'Динамика RMSE и остатков во время обучения')
    x_label : str, optional
        Подпись оси X (по умолчанию 'Эпоха')
    y_label : str, optional
        Подпись оси Y (по умолчанию 'RMSE')
    figsize : tuple, optional
        Размер графика (по умолчанию (13, 6))
    train_alpha : float, optional
        Прозрачность линии RMSE для обучающей выборки (по умолчанию 0.7)
    test_alpha : float, optional
        Прозрачность линии RMSE для тестовой выборки (по умолчанию 0.7)
    residual_alpha : float, optional
        Прозрачность линии остатков (по умолчанию 0.7)
    show_stats : bool, optional
        Показывать ли статистику RMSE и остатков (по умолчанию True)
    label_angle : int, optional
        Угол наклона подписей по оси X (по умолчанию 0)

    Примечания:
    -----------
    - График отображает RMSE для обучающей и тестовой выборок, а также остатки (RMSE_test - RMSE_train) в виде линий.
    - Автоматически настраиваются пределы осей и шаг подписей для больших данных.
    - При show_stats выводится статистика по rmse_train, rmse_test и остаткам.
    """
    if not isinstance(history_df, pd.DataFrame):
        raise TypeError("history_df должен быть типа pd.DataFrame")

    required_columns = ['epoch', 'rmse_train', 'rmse_test']
    if not all(col in history_df.columns for col in required_columns):
        raise ValueError("history_df должен содержать столбцы: epoch, rmse_train, rmse_test")

    epochs = history_df['epoch'].values
    rmse_train = history_df['rmse_train'].values
    rmse_test = history_df['rmse_test'].values
    residuals = rmse_test - rmse_train

    fig, ax = plt.subplots(figsize=figsize)

    # строим линии для rmse_train, rmse_test и residuals
    ax.plot(epochs, rmse_train, label='RMSE (train)', alpha=train_alpha, 
            color='skyblue', linewidth=2)
    ax.plot(epochs, rmse_test, label='RMSE (test)', alpha=test_alpha, 
            color='lightcoral', linewidth=2)
    ax.plot(epochs, residuals, label='Остатки (test - train)', alpha=residual_alpha, 
            color='limegreen', linewidth=2)

    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title, fontsize=14, pad=20)
    ax.legend()

    # настраиваем ось X
    n_epochs = len(epochs)
    if n_epochs > 30:
        step = max(1, n_epochs // 20)
        ax.set_xticks(epochs[::step])
        ax.set_xticklabels(epochs[::step], rotation=label_angle)
    else:
        ax.set_xticks(epochs)
        ax.set_xticklabels(epochs, rotation=label_angle)

    # добавим сеточку
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.grid(axis='x', linestyle='--', alpha=0.7)
    ax.set_axisbelow(True)

    # автоматом настраиваем пределы
    y_min = min(rmse_train.min(), rmse_test.min(), residuals.min())
    y_max = max(rmse_train.max(), rmse_test.max(), residuals.max())
    y_range = y_max - y_min
    ax.set_ylim(y_min - 0.1 * y_range, y_max + 0.1 * y_range)

    plt.tight_layout()
    plt.show()

    # статистика
    if show_stats:
        stats_df = pd.DataFrame({
            'RMSE (train)': rmse_train,
            'RMSE (test)': rmse_test,
            'Остатки (test - train)': residuals
        })
        display(stats_df.describe())
````

### Функция ```hyper_search```

````python
def hyper_search(X_train, y_train, X_test, y_test, param_grid, num_epochs, print_every):
    """
    Функция для выполнения поиска гиперпараметров нейронной сети по заданной сетке параметров
    с целью минимизации RMSE на тестовом наборе данных. Сохраняет историю обучения для всех комбинаций.

    Параметры:
    ----------
    X_train : array-like
        Обучающий набор признаков (pd.DataFrame, np.ndarray)
    y_train : array-like
        Обучающий набор целевых значений (pd.Series, np.ndarray, list)
    X_test : array-like
        Тестовый набор признаков (pd.DataFrame, np.ndarray)
    y_test : array-like
        Тестовый набор целевых значений (pd.Series, np.ndarray, list)
    param_grid : dict
        Сетка гиперпараметров для перебора (словарь с ключами и списками значений)
    num_epochs : int
        Количество эпох для обучения модели
    print_every : int
        Частота вывода информации о процессе обучения (в эпохах)

    Возвращает:
    -----------
    best_model : NNet
        Обученная модель с лучшими гиперпараметрами
    best_params : dict
        Словарь с лучшими найденными гиперпараметрами
    history_df : pd.DataFrame
        DataFrame с историей обучения для всех комбинаций параметров, содержит столбцы:
        - combo_idx: индекс комбинации параметров
        - params: строка с параметрами
        - epoch: номер эпохи
        - loss: значение функции потерь на обучающем наборе
        - rmse_train: RMSE на обучающем наборе
        - rmse_test: RMSE на тестовом наборе
    """
    best_rmse = float('inf')
    best_params = None
    best_model = None
    history_list = []

    # создаем все комбинации параметров по сетке
    keys, values = zip(*param_grid.items())
    total_combinations = np.prod([len(v) for v in values])
    print(f"Кол-во комбинаций: {total_combinations}")

    for combo_idx, combo in enumerate(product(*values)):
        params = dict(zip(keys, combo))
        print(f"\n[{combo_idx+1}/{total_combinations}] Считаем: {params}")

        # создаем архитектуру нейросети
        input_size = X_train.shape[1]
        if 'n_hidden_layers' in params:
            hidden_layers = [params['n_neurons']] * params['n_hidden_layers']
            n_neurons_list = [input_size] + hidden_layers + [1]
        else:
            n_neurons_list = [input_size, 8, 6, 1]

        # при необходимости передаем параметры в функцию активации
        act_params = {}
        if params['activation'] == 'LeakyReLU':
            act_params = {'alpha': params.get('leaky_alpha', 0.01)}

        # создаем и тренируем модель
        model = NNet(
            n_neurons_list,
            activation=params['activation'],
            dropout_rate=params['dropout_rate'],
            use_batchnorm=params['use_batchnorm'],
            activation_params=act_params
        )

        # предполагаем, что train_model возвращает историю обучения
        history = model.train_model(
            X_train, y_train, X_test, y_test,
            learning_rate=params['learning_rate'],
            num_epochs=num_epochs,
            batch_size=params['batch_size'],
            print_every=print_every
        )

        # получаем итоговую метрику
        metrics = model.get_metrics(X_test, y_test)
        test_rmse = metrics['RMSE']

        print(f"RMSE (test): {test_rmse:.4f}")

        for _, row in history.iterrows():
            history_list.append({
                'combo_idx': combo_idx + 1,
                'params': str(params),
                'epoch': row['epoch'],
                'loss': row['loss'],
                'rmse_train': row['rmse_train'],
                'rmse_test': row['rmse_test']
            })

        if test_rmse < best_rmse:
            best_rmse = test_rmse
            best_params = params
            best_model = model

    print("Лучшие параметры модели:")
    for key, value in best_params.items():
        print(f"{key}: {value}")
    print(f"Лучший показатель RMSE: {best_rmse:.4f}")

    # создаем DataFrame из истории
    history_df = pd.DataFrame(history_list)
    return best_model, best_params, history_df
````

### Функция ```plot_fact_forecast_comparison```

````python
def plot_fact_forecast_comparison(
    y_true,
    y_test_predicted_b,
    y_test_predicted_bl,
    title: str = 'Сравнение факта, прогнозов и остатков (лучшая и базовая модели)',
    x_label: str = 'Наблюдения',
    y_label: str = 'Значения',
    residual_label: str = 'Остатки (прогноз - факт)',
    figsize: tuple = (13, 6),
    fact_alpha: float = 0.7,
    forecast_b_alpha: float = 0.5,
    forecast_bl_alpha: float = 0.5,
    residual_b_alpha: float = 0.7,
    residual_bl_alpha: float = 0.7,
    show_stats: bool = True,
    label_angle: int = 0
):
    """
    Функция для визуального сравнения фактических значений, прогнозов двух моделей
    (лучшей и базовой) в виде столбцов и их остатков в виде линий.

    Параметры:
    ----------
    y_true : array-like
        Фактические значения (pd.Series, np.ndarray, list)
    y_test_predicted_b : array-like
        Прогнозные значения лучшей модели (pd.Series, pd.DataFrame, np.ndarray, list)
    y_test_predicted_bl : array-like
        Прогнозные значения базовой модели (pd.Series, pd.DataFrame, np.ndarray, list)
    title : str, optional
        Заголовок графика (по умолчанию 'Сравнение факта, прогнозов и остатков (лучшая и базовая модели)')
    x_label : str, optional
        Подпись оси X (по умолчанию 'Наблюдения')
    y_label : str, optional
        Подпись оси Y для фактических и прогнозных значений (по умолчанию 'Значения')
    residual_label : str, optional
        Подпись оси Y для остатков (по умолчанию 'Остатки (прогноз - факт)')
    figsize : tuple, optional
        Размер графика (по умолчанию (13, 6))
    fact_alpha : float, optional
        Прозрачность столбцов факта (по умолчанию 0.7)
    forecast_b_alpha : float, optional
        Прозрачность столбцов прогноза лучшей модели (по умолчанию 0.5)
    forecast_bl_alpha : float, optional
        Прозрачность столбцов прогноза базовой модели (по умолчанию 0.5)
    residual_b_alpha : float, optional
        Прозрачность линии остатков лучшей модели (по умолчанию 0.7)
    residual_bl_alpha : float, optional
        Прозрачность линии остатков базовой модели (по умолчанию 0.7)
    show_stats : bool, optional
        Показывать ли статистику сравнения (по умолчанию True)
    label_angle : int, optional
        Угол наклона подписей по оси X (по умолчанию 0)

    Примечания:
    -----------
    - График отображает фактические и прогнозные значения (лучшая и базовая модели) как столбцы,
      а остатки (прогноз - факт) как линии на вторичной оси Y.
    - Автоматически настраиваются пределы осей и шаг подписей для больших данных.
    - При show_stats выводится статистика по фактическим, прогнозным значениям и остаткам.
    """
    # проверка входных данных
    if not isinstance(y_true, (pd.Series, pd.DataFrame, np.ndarray, list)):
        raise TypeError("y_true должен быть типа pd.Series, pd.DataFrame, np.ndarray или list")
    if not isinstance(y_test_predicted_b, (pd.Series, pd.DataFrame, np.ndarray, list)):
        raise TypeError("y_test_predicted_b должен быть типа pd.Series, pd.DataFrame, np.ndarray или list")
    if not isinstance(y_test_predicted_bl, (pd.Series, pd.DataFrame, np.ndarray, list)):
        raise TypeError("y_test_predicted_bl должен быть типа pd.Series, pd.DataFrame, np.ndarray или list")

    # преобразуем в numpy массивы
    y_true = np.array(y_true)
    y_test_predicted_b = np.array(y_test_predicted_b)
    y_test_predicted_bl = np.array(y_test_predicted_bl)

    # проверка совпадения размеров
    if not (len(y_true) == len(y_test_predicted_b) == len(y_test_predicted_bl)):
        raise ValueError("Размеры y_true, y_test_predicted_b и y_test_predicted_bl должны совпадать")

    # вычисляем остатки
    residuals_b = y_test_predicted_b - y_true
    residuals_bl = y_test_predicted_bl - y_true
    
    n_observations = len(y_true)
    x_index = np.arange(n_observations)

    # создаем фигуру и основную ось
    fig, ax1 = plt.subplots(figsize=figsize)

    # строим столбцы для факта
    ax1.bar(x_index - 0.2, y_true, width=0.3, label='Факт', alpha=fact_alpha, 
            color='skyblue', edgecolor='navy', linewidth=1.2)

    # строим столбцы для прогноза лучшей модели
    ax1.bar(x_index, y_test_predicted_b, width=0.3, label='Прогноз (лучшая)', 
            alpha=forecast_b_alpha, color='salmon', edgecolor='darkred', linewidth=1.2)

    # строим столбцы для прогноза базовой модели
    ax1.bar(x_index + 0.2, y_test_predicted_bl, width=0.3, label='Прогноз (базовая)', 
            alpha=forecast_bl_alpha, color='purple', edgecolor='darkred', linewidth=1.2)

    # настраиваем основную ось Y
    ax1.set_xlabel(x_label)
    ax1.set_ylabel(y_label)
    ax1.set_title(title, fontsize=14, pad=20)

    # создаем вторичную ось Y для остатков
    ax2 = ax1.twinx()
    ax2.plot(x_index, residuals_b, label='Остатки (лучшая)', alpha=residual_b_alpha, 
             color='limegreen', linewidth=2, marker='^', markersize=4)
    ax2.plot(x_index, residuals_bl, label='Остатки (базовая)', alpha=residual_bl_alpha, 
             color='darkgreen', linewidth=2, marker='s', markersize=4)
    ax2.set_ylabel(residual_label)

    # объединяем легенды обеих осей
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + lines2, loc='best')

    # настраиваем ось X
    if n_observations > 30:
        step = max(1, n_observations // 20)
        ax1.set_xticks(x_index[::step])
        ax1.set_xticklabels(x_index[::step], rotation=label_angle)
    else:
        ax1.set_xticks(x_index)
        ax1.set_xticklabels(x_index, rotation=label_angle)

    # фигачим сеточку
    ax1.grid(axis='y', linestyle='--', alpha=0.7)
    ax1.set_axisbelow(True)

    # автоматическая настройка пределов для основной оси
    y_min = min(y_true.min(), y_test_predicted_b.min(), y_test_predicted_bl.min())
    y_max = max(y_true.max(), y_test_predicted_b.max(), y_test_predicted_bl.max())
    y_range = y_max - y_min
    ax1.set_ylim(y_min - 0.1 * y_range, y_max + 0.1 * y_range)

    # автоматическая настройка пределов для вторичной оси
    res_min = min(residuals_b.min(), residuals_bl.min())
    res_max = max(residuals_b.max(), residuals_bl.max())
    res_range = res_max - res_min
    ax2.set_ylim(res_min - 0.1 * res_range, res_max + 0.1 * res_range)

    plt.tight_layout()
    plt.show()

    # статистика
    if show_stats:
        stats_df = pd.DataFrame({
            'Факт': y_true,
            'Прогноз (лучшая)': y_test_predicted_b,
            'Прогноз (базовая)': y_test_predicted_bl,
            'Остатки (лучшая)': residuals_b,
            'Остатки (базовая)': residuals_bl,
            'Абс. ошибка (лучшая)': np.abs(residuals_b),
            'Абс. ошибка (базовая)': np.abs(residuals_bl)
        })

        display(stats_df.describe())

        # метрики качества для лучшей модели
        mae_b = np.mean(np.abs(residuals_b))
        mse_b = np.mean(residuals_b ** 2)
        rmse_b = np.sqrt(mse_b)
        mape_b = np.mean(np.abs(residuals_b / y_true)) * 100 if np.all(y_true != 0) else np.nan

        # метрики качества для базовой модели
        mae_bl = np.mean(np.abs(residuals_bl))
        mse_bl = np.mean(residuals_bl ** 2)
        rmse_bl = np.sqrt(mse_bl)
        mape_bl = np.mean(np.abs(residuals_bl / y_true)) * 100 if np.all(y_true != 0) else np.nan

        print("Метрики для лучшей модели:")
        print(f"RMSE: {rmse_b:.4f}")
        print("\nМетрики для базовой модели:")
        print(f"RMSE: {rmse_bl:.4f}")
````

#### Признак ```abs_mag```

**Комментарий**: подробнее рассмотрим признак ```abs_mag``` -- абсолютная звёздная величина.

Абсолютная звёздная величина (M) для звёзд определяется как видимая звёздная величина объекта, если бы он был расположен на расстоянии 10 парсек от наблюдателя и не испытывал бы ни межзвёздного, ни атмосферного поглощения. 

Если известна видимая звёздная величина $m$ и расстояние до объекта $d$, можно вычислить абсолютную звёздную величину по формуле:
$M = m - 5\lg {\frac {d}{d_{0}}}$,
где $d_{0}= 10$ пк $\approx 32.62$ световых лет. 
Этой формуле эквивалентны 
$M = m + 5 - 5\lg r$ и $M = m + 5 + 5 \lg \pi^{\prime\prime}$, где 
$r$ и $\pi^{\prime \prime}$ -- расстояние до звезды в парсеках и годичный параллакс в секундах соответственно.

Абсолютная звёздная величина связана со светимостью следующим соотношением:

$\lg {\frac {L}{L_{\odot }}}=0,4(M_{\odot }-M)$,
где $L_{\odot}$ и $M_{\odot}$ -- светимость и абсолютная звёздная величина Солнца.

## <u>Подготовка данных</u>

**Комментарий**: выполним теперь подготовку данных для последующего обучения и проверки нейронной сети.

````python
# преобразуем данные
X_train_p = preprocessor.fit_transform(X_train)
X_test_p = preprocessor.transform(X_test)
````

````python
# выведем размерность после преобразования
print(f"Размерность тренировочной выборки: {X_train_p.shape}")
print(f"Размерность тестовой выборки: {X_test_p.shape}")
````

````python
# преобразуем данные теперь в тензоры
X_train_t = torch.FloatTensor(X_train_p)
X_test_t = torch.FloatTensor(X_test_p)
y_train_t = torch.FloatTensor(y_train.values)
y_test_t = torch.FloatTensor(y_test.values)
````

````python
# выведем размерности полученных выборок
print("Размерность тренировочной выборки:", X_train_t.size())
print("Размерность тестовой выборки:", X_test_t.size())
print("Размерность тренировочной выборки целевого признака:", y_train_t.size())
print("Размерность тестовой выборки целевого признака:", y_test_t.size())
````

## <u>Построение базовой нейронной сети</u>

**Комментарий**: теперь необходимо создать класс для задания архитектуры нейронной сети, выбрать количество скрытых слоёв, количество нейронов на них, функции активации на скрытых и выходном слоях. Затем необходимо создать функцию для обучения нейронной сети и выполнить обучение модели.

Важно отметить, что задача является задачей регрессии, где нам необходимо предсказать температуру поверхности звезды (1 выходной нейрон), в качестве метрики необходимо использовать $RMSE$ (СКО).

По результатам обучения, необходимо построить график «Факт-Прогноз», где по горизонтальной оси будут отложены условные номера звёзд, а по вертикальной -- температура поверхности в Кельвинах.

**Комментарий**: в качестве базовой модели возьмем модель с двумя скрытыми слоями (8 и 6 нейронов), 1 выходной нейрон, количество входов сети равно количеству признаков. Функцию активации выберем $ReLU$. Пока без $Dropout$ и без батчей. Также используем $Adam$ оптимизатор c ```learning_rate=1e-3```.

````python
# создадим и обучим базовую модель 
n_neurons = [X_train_t.shape[1], 8, 6, 1]
model = NNet(n_neurons, activation='ReLU', dropout_rate=0.0)

baseline_train_result = model.train_model(
    X_train_t, y_train_t, X_test_t, y_test_t,
    learning_rate = 1e-3,
    num_epochs = NUM_EPOCHS,
    batch_size = 0,
    optimizer_type = 'Adam',
    print_every = PRINT_EVERY
)

# предсказываем и получаем метрику
predictions = model.predict(X_test_t)
metrics = model.get_metrics(X_test_t, y_test_t)
print(f"Итоговая метрика RMSE: {metrics['RMSE']:.4f}")
````

````python
# теперь получим предсказание на тестовых данных и построим график факт-прогноз
model.eval()

test_preds_final_bl = model.forward(X_test_t).flatten()
y_test_actual_bl = y_test_t.detach().cpu().numpy()
y_test_predicted_bl = test_preds_final_bl.detach().cpu().numpy()

# строим график факт-прогноз
plot_fact_forecast(
    y_true=y_test_actual_bl,
    y_pred=y_test_predicted_bl,
    title=f'Сравнение фактического и прогнозного значения температуры звезды',
    x_label='Номер звезды',
    y_label='Температура, (К)'
)

# строим график для сравнения метрик
plot_training_history(
    baseline_train_result,
    title = 'Динамика RMSE во время обучения',
    x_label = 'Эпоха',
    y_label = 'RMSE',
    show_stats = True
)
````

````python
# выведем архитектуру модели
print(model.get_architecture_info())
````

**Промежуточный вывод**: была построена модель с архитектурой: **14** входов, **2** скрытых слоя (**8** и **6** нейронов), ф-ция активации $ReLU$, без батчей и $Dropout$, использовался оптимизатор $Adam$ с ```learning_rate=1e-3```. Использовалось **30000** эпох при обучении. По резлультатам получена метрика $RMSE=3071$. Модель обучивается на $\sim$**22000** эпохе, дальше ```residuals``` $RMSE(train)-RMSE(test)$ начинают расти. Перейдем теперь к улучшению нейросети.
